- アメダス地点の緯度，経度を含む表
- 地名と経度緯度を含む表

以上からアメダス地点に最も近い地名を割り当てる

```python
def calc_name(latitude,longitude,yubin):
    min_norm = 1e9
    min_name = 'unknown'
    for y_name,y_latitude,y_longitude in zip(yubin['name'],yubin['latitude'],yubin['longitude']):
        norm = (latitude - y_latitude)**2 + (longitude - y_longitude)**2
        if min_norm > norm:
            min_norm = norm
            min_name = y_name
    
    return min_name, min_norm
            
```

```python
amedas = pd.read_csv('all3.csv', sep=',')
print(amedas)
```

```python
yubin = pd.read_table('yubin_jusho.tsv', sep='\t')
print(yubin)
```

```python
location = []
distance = []

for station,latitude,longitude in zip(amedas['station'],amedas['latitude'],amedas['longitude']):
    print(station, latitude, longitude)
    result = calc_name(latitude,longitude,yubin)
    print('=> ', result[0])
    location.append(result[0])
    distance.append(result[1])
```

```python
amedas = amedas.assign(location=location, distance=distance)

print(amedas)
```
